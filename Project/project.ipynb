```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('지하철_전처리.csv', encoding= 'cp949')
df
```

```python
import json
geo_str = json.load(open('skorea_municipalities_geo_simple.json',
                         encoding='utf8'))
```

```python
df1 = df[df['호선']=='01호선']
```

```python
map = folium.Map(location=[df1['위도'].mean(), df1['경도'].mean()], tiles='cartodb positron', zoom_start=11)

map.choropleth(
    geo_data = geo_str,
    fill_color = 'snow')


for i in df1.index:
    folium.Circle(
        location=[df1['위도'][i], df1['경도'][i]],
        color='blue'
    ).add_to(map)
map
```
